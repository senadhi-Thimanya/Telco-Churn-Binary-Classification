```python
import joblib
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
```

```python
dt = joblib.load('/kaggle/input/telco-train-test-split-dataset/dt_results.pkl')
nn = joblib.load('/kaggle/input/telco-train-test-split-dataset/nn_results.pkl')

fpr_dt, tpr_dt, _ = roc_curve(dt['y_true'], dt['y_pred_proba'])
fpr_nn, tpr_nn, _ = roc_curve(nn['y_true'], nn['y_pred_proba'])

plt.figure(figsize=(9,7))
plt.plot(fpr_dt, tpr_dt, label=f"Decision Tree  (AUC = {auc(fpr_dt, tpr_dt):.4f})", linewidth=2.5)
plt.plot(fpr_nn, tpr_nn, label=f"Neural Network (AUC = {auc(fpr_nn, tpr_nn):.4f})", linewidth=2.5)
plt.plot([0,1], [0,1], 'k--', label="Random classifier")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison  Decision Tree vs Neural Network')
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.savefig('/kaggle/working/roc_curve.png', dpi=300, bbox_inches='tight')
plt.show()
```
